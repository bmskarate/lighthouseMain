```python
import pandas as pd
import numpy as np
```

```python
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
                         'foo', 'foo', 'qux', 'qux'],
                        ['one', 'two', 'one', 'two',
                         'one', 'two', 'one', 'two']]))
```

```python
_index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
```

```python
df = pd.DataFrame(np.random.randn(8, 2), index=_index, columns=['A', 'B'])
```

```python
df
```

```python
df1 = pd.DataFrame(np.random.randn(8, 2), index=_index, columns=['A', 'B'])
df1
```

```python
df2 = df[:4]
df2
```

## pivoting tables

```python
df3 = pd.DataFrame({'A': ['one', 'one', 'two', 'three'] * 3,
                       'B': ['A', 'B', 'C'] * 4,
                       'C': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                       'D': np.random.randn(12),
                       'E': np.random.randn(12)}) # how to len(A)?
```

```python
df3
```

```python
pd.pivot_table(df3, values='D', index=['A', 'B'], columns=['C'])
```
