# 4) Create a dataframe with the missing names filled in.

```python
empty_df = json_normalize(data=nested_json, record_path="mjtheme_namecode")
empty_df
```

```python
dict_for_empty_df = {1:"Economic management", 2:"Economic management", 3:"Rule of law", 4:"Financial and private sector development",
          5:"Trade and integration", 6:"Social protection and risk management", 7: "Social dev/gender/inclusion", 8: "Human development",
          9:"Urban development",10:"Rural development", 11:"Environment and natural resources management"
         }
```

```python
empty_df['name'] = empty_df['code'].apply(lambda x:dict_for_empty_df.get(int(x)))
```

```python
empty_df.groupby("name").count()
```

```python
active_projects["sector_dict"]
```

```python
active_projects["sector_name"] = active_projects["sector_dict"].apply(get_name)
```
