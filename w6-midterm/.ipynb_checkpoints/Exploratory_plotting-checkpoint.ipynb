```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("data/flights_2019_jan_nov_dec_CLEAN.csv")
```

```python
df.head()
```

# Plot total Flights

```python
# total flights
total_flights = pd.DataFrame()
total_flights["airline_names"] = df.airline_names
total_flights["flights"] = df.flights
```

```python
total_flights_plot = pd.DataFrame(total_flights.groupby(by="airline_names").count())
```

```python
total_flights_plot.reset_index(inplace=True)
total_flights_plot = total_flights_plot.sort_values(by="flights", ascending=False)
total_flights_plot
```

```python
import pandas as pd
from matplotlib import pyplot as plt
 
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(total_flights_plot.airline_names, 
        total_flights_plot.flights,
        color=("b","r","b","b","b","b","b","b","b"))
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# remove ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 16, fontweight ='bold',
             color = 'grey')
 
# Add Plot Title
ax.set_title('Total Number of Flights',
             loc ='left',
             fontsize = 24)
plt.yticks(fontsize=20)
# Add Text watermark
fig.text(0.9, 0.15, 'Lrossi', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
fig.savefig('total_number_flights.png', bbox_inches='tight', dpi=fig.dpi)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.pie(total_flights_plot.flights,
        labels=total_flights_plot.airline_names,
        radius=2,
        shadow=True,
        labeldistance=1.2,
        explode=[0,0.4,0,0,0,0,0,0,0,0])

plt.title('Total Flights')

plt.axis('equal')

plt.savefig('total_flights_pie.png', dpi=fig.dpi)
plt.show() 
```

## Plot total delayed flights

```python
# total delayed flights
total_delayed_flights = pd.DataFrame()
total_delayed_flights = df[df["dep_delay"] > 0]
total_delayed_flights = total_delayed_flights.filter(['flights', 'airline_names'])
total_delayed_flights = total_delayed_flights.groupby(by="airline_names").sum()
```

```python
total_delayed_flights = total_delayed_flights.sort_values(by="flights", ascending=False)
total_delayed_flights.reset_index(inplace=True)
total_delayed_flights
```

```python
import pandas as pd
from matplotlib import pyplot as plt
 
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(total_delayed_flights.airline_names, 
        total_delayed_flights.flights,
       color=("b","b","b","r","b","b","b","b","b"))
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# remove ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 16, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('Total Number of Delayed Flights',
             loc ='left',
             fontsize = 24)
 
plt.yticks(fontsize=20)
# Add Text watermark
fig.text(0.9, 0.15, 'Lrossi', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
fig.savefig('total_delayed_flights.png', bbox_inches='tight', dpi=fig.dpi)
plt.show()
```

# On time vs delayed

```python
ontime_vs_delay = pd.DataFrame()
```

```python
ontime_vs_delay["delay_total"] = df[df["dep_delay"] > 0].count()
```

```python
ontime_vs_delay["ontime_total"] = df[df["dep_delay"] <= 0].count()
```

```python
ontime_vs_delay.reset_index(inplace=True)
```

```python
ontime_vs_delay = ontime_vs_delay.drop(columns="index")
```
